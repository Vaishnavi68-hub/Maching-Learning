```python
import pandas as pd
```

```python
s  = pd.Series([1,2,3,4,5])
print(s)
print(type(s))


#ACCESSING

#indexing

print(s[0])
print(s[3])
s[0] =100

print(s)


```

```python
#Creating dataframes through 

#Dictonaries

info = {
    "Name" : ["alice" , "bob" , "charlie"],
    "Age" : [12,23,34],
    "CGPA" : [9.1,2.3,9.9]
}

df = pd.DataFrame(info)
print(df)
print(df.index)
print(df.columns)
```

```python
#Lists

df = pd.DataFrame([["Alice" , 23] ,["Bob" , 32] ,["Eve" , 43]] , columns=["Name" ,"Age"])
print(df)
```

```python
# Creating dataframes from numpy array

import numpy as np

np_arr = np.array([[1,2,3] , [4,5,6]])

df = pd.DataFrame(np_arr , columns=["a" , "b" , "c"])

print(df)
```
