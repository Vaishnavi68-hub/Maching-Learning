```python
#Creating dataframes through 

#Dictonaries

info = {
    "Name" : ["alice" , "bob" , "charlie"],
    "Age" : [12,23,34],
    "CGPA" : [9.1,2.3,9.9]
}

df = pd.DataFrame(info)
print(df)
print(df.index)
print(df.columns)
```
