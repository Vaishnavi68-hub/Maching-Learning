```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
arr1 = np.array([2,5,4,3,2])
print(arr ** 2)
print(arr + 10)
print(arr + arr1)
```

```python
#broadcasting


arr = np.array([1,2,3,4,5])
arr2 = np.array([[1,2,3,4,5] ,[4,5,6,3,2]])
print(arr + 5)
print(arr + arr2)
print(arr2,arr2.shape)
```

```python
#normalisation

arr = np.array([[1,2] ,[3,4]])
mean = np.mean(arr)
std_dev = np.std(arr)

normalized_arr = (arr - mean)/std_dev
print(normalized_arr)


print(mean)
print(std_dev)
```

```python
# Mathematical functions

# Aggregation

arr = np.array([1,2,3,4,5,])

print(np.sum(arr))
print(np.prod(arr))
print(np.min(arr))
print(np.max(arr))
print(np.argmax(arr)) #index of max value
print(np.argmin(arr)) # index of min value
print(np.mean(arr))
print(np.std(arr))
print(np.median(arr))
print(np.var(arr))



#Power 

print(np.square(arr))
print(np.sqrt(arr))


#Log Function
print(np.log(arr))
print(np.log10(arr))
print(np.exp(arr))

```

```python
#rounding 


print(np.round(3.14))
print(np.ceil(3.14))
print(np.floor(3.14))
print(np.trunc(3.14))
```

```python
arr = np.array([1,2,3,5,-6,4,2,4,5,])

print(np.unique(arr))
print(np.sort(arr))


print(np.abs(arr))  #makes all values positive and return
```
