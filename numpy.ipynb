```python
import numpy as np
import time
```

```python
#execution performance
size = 1_000_000

#measure time
#python list
py_list = list(range(size))

start = time.time() #current time

sq_list = [x**2 for x in py_list]

end = time.time() #end time

print("python list time:" , end-start ,"seconds")



#numpy  arrays
np_arr = np.array(py_list)
start = time.time()

sq_array = np_arr ** 2
end = time.time()

print("python list time:" , end-start ,"seconds")

```

```python
#memory

import sys

print("Size of python list =" , sys.getsizeof(py_list)  , "bytes")


print("Size of numpy array = ", sys.getsizeof(np_arr.nbytes) , "bytes")

```
