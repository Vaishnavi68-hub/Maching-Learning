```python
import numpy as np
```

```python
# Linear regression

class LinearRegression:
    def __init__(self ,learning_rate=0.01 , n_iter = 1000):
        self.bias = None
        self.weights = None
        self.lr = learning_rate
        self.n_iter = n_iter

    def fit(self , X , y): # X_train , y_train
        m , n = X.shape #(number of samples (m), no of features(n) -> rows and columns)


        #Step 1 -- initailize the parameters
        self.bias = 0
        self.weights = np.zeros(n) # [ 0,0,0....]


        #Gradient descent

        for i in range(self.n_iter):
            #Step 2 - calcutlate y_pred
            y_pred = self.bias + np.dot(X, self.weights)



            #Step 3 - gradient calcuttion
            db = (1/m) * np.sum(y_pred - y)
            dw = (1/m) * np.dot(X.T ,(y_pred - y))



            #Step 4 - convergence theorem - upadate parameters
            self.bias = self.bias - self.lr * db
            self.weights = self.weights - self.lr * db
        




    def predict(self , X):
        y_pred = self.bias + np.dot(X, self.weights)
        return y_pred

        
        

         

        



    
```

```python
X = np.array([[1],[2],[3],[4],[5]])
y = np.array([2,4,6,8,10])


model = LinearRegression()
model.fit(X,y)
y_pred = model.predict(X)

print(y_pred)
```
